`````````python
cnt_temp_am_pm = 0;
for i in data:
    if i == '-':
        break
    else:
        cnt_temp_am_pm += 1
        
cnt_temp_am_pm -= 3
cnt_temp_am_pm
`````````

`````````python
am_pm = data[cnt_temp_am_pm] + data[cnt_temp_am_pm + 1]
am_pm
`````````

`````````python
#                09  /    09  /   2019 ,     10   :  29    pm
if am_pm == 'am' or am_pm == 'pm' or am_pm == 'AM' or am_pm == 'PM' or am_pm == 'Am' or am_pm == 'Pm':
    pattern = "\d{1,2}\/\d{2,4}\/\d{2,4},\s\d{1,2}:\d{1,2}\s\w{1,2}\s-\s"    # for 12 hr format
else:
    pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
`````````

`````````python
messages = re.split(pattern, data)[1:]
messages
`````````

`````````python
dates = re.findall(pattern, data)
dates
`````````

`````````python
time_format = dates[0]
time_format
`````````

`````````python
cnt = 2
newTimeFormat = ""
for i in time_format:
    if i == '/':
        cnt -= 1
    if i == ',' :
        break;
    if cnt == 0:
        newTimeFormat += i;

newTimeFormat = newTimeFormat[1:]
print(newTimeFormat)
print(time_format)
`````````

`````````python
# seperate users and messages 
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\w]+?):\s', message)
    if(entry[1:]): # user name
        users.append(entry[0] + entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
        
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace = True)

df.head()
`````````
